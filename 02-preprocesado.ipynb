```python
import pandas as pd
import numpy as np
import os
```

```python
if os.path.exists('train.csv'):
    ruta = 'train.csv'
else:
    ruta = 'sample_data/train.csv'

datos = pd.read_csv(ruta)
print('Datos cargados correctamente.')
print(f'Total de filas: {len(datos)}')
print(f'Total de columnas: {len(datos.columns)}')
```

```python
datos.head()
```

```python
datos.info()
```

```python
# 1. IDENTIFICAMOS LOS VALORES FALTANTES
print('Valores faltantes por columna:')
valores_faltantes = datos.isnull().sum()
valores_faltantes = valores_faltantes[valores_faltantes > 0]

if len(valores_faltantes) > 0:
    print(valores_faltantes)
else:
    print('No hay valores faltantes')
```

```python
# 2. ASIGNAMOS DATOS EN LOS VALORES FALTANTES
# Hacemos una copia para no modificar los datos originales
datos_procesados = datos.copy()

# Para columnas numéricas: rellenar con la mediana
columnas_numericas = datos_procesados.select_dtypes(include=['int64', 'float64']).columns
for columna in columnas_numericas:
    if datos_procesados[columna].isnull().sum() > 0:
        mediana = datos_procesados[columna].median()
        datos_procesados[columna].fillna(mediana, inplace=True)
        print(f'Rellenados {columna} con mediana: {mediana}')

# Para columnas categóricas: rellenar con la moda (valor más frecuente)
columnas_categoricas = datos_procesados.select_dtypes(include=['object']).columns
columnas_categoricas = [col for col in columnas_categoricas if col != 'RENDIMIENTO_GLOBAL']

for columna in columnas_categoricas:
    if datos_procesados[columna].isnull().sum() > 0:
        moda = datos_procesados[columna].mode()[0]
        datos_procesados[columna].fillna(moda, inplace=True)
        print(f'Rellenados {columna} con moda: {moda}')

print('\n Asignación completada')
```

```python
# 3. IDENTIFICAMOS LOS TIPOS DE VARIABLES
print('Variables numéricas:')
numericas = datos_procesados.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(numericas)

print('\nVariables categóricas:')
categoricas = datos_procesados.select_dtypes(include=['object']).columns.tolist()
# Excluir ID y la variable objetivo
categoricas = [col for col in categoricas if col not in ['ID', 'RENDIMIENTO_GLOBAL']]
print(categoricas)
```

```python
# 4. CONVERTIMOS LAS VARIABLES CATEGÓRICAS A ONE-HOT ENCODING
# One-hot encoding convierte categorías en columnas binarias (0 o 1)
# Ejemplo: si 'sexo' tiene valores M/F, se crean columnas sexo_M y sexo_F

print('Aplicando One-Hot Encoding...')
print(f'Columnas antes: {len(datos_procesados.columns)}')

# Aplicar one-hot encoding a las variables categóricas
datos_procesados = pd.get_dummies(datos_procesados, columns=categoricas, drop_first=True)

print(f'Columnas después: {len(datos_procesados.columns)}')
print('One-Hot Encoding completado')
```

```python
# 5. NORMALIZAMOS LAS VARIABLES NUMÉRICAS
# La normalización escala los valores entre 0 y 1
# Útil para que todas las variables tengan la misma escala

from sklearn.preprocessing import MinMaxScaler

# Excluir ID de la normalización
columnas_a_normalizar = [col for col in numericas if col != 'ID']

if len(columnas_a_normalizar) > 0:
    scaler = MinMaxScaler()
    datos_procesados[columnas_a_normalizar] = scaler.fit_transform(datos_procesados[columnas_a_normalizar])
    print(f'Normalizadas {len(columnas_a_normalizar)} columnas numéricas')
    print('Valores ahora están entre 0 y 1')
else:
    print('No hay columnas numéricas para normalizar')
```

```python
# 6. VERIFICAMOS EL RESULTADO FINAL
print('Resumen del dataset procesado:')
print(f'Filas: {len(datos_procesados)}')
print(f'Columnas: {len(datos_procesados.columns)}')
print(f'\nValores faltantes restantes: {datos_procesados.isnull().sum().sum()}')

# Ver las primeras filas del dataset procesado
datos_procesados.head()
```

```python
# 7. GUARDAMOS LOS DATOS PROCESADOS
# Guardar el dataset limpio para usarlo en el modelo
datos_procesados.to_csv('train_procesado.csv', index=False)
print('Datos procesados guardados en: train_procesado.csv')
```
